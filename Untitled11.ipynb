```python
# Calculate 6-month and 12-month simple moving averages
eth_dat['6-month-SMA'] = eth_dat['price'].rolling(window=6).mean()
eth_dat['12-month-SMA'] = eth_dat['price'].rolling(window=12).mean()


# Plot the original price and SMAs
eth_dat.plot(x='TIME', y=['price', '6-month-SMA', '12-month-SMA'])
plt.show()
```

```python
# Simple Moving Averages (SMA)
eth_trend['6-month-SMA'] = eth_trend['price'].rolling(window=6).mean()
eth_trend['12-month-SMA'] = eth_trend['price'].rolling(window=12).mean()
eth_trend['600-day-SMA'] = eth_trend['price'].rolling(window=600).mean()
eth_trend['1200-day-SMA'] = eth_trend['price'].rolling(window=1200).mean()

# Exponentially Weighted Moving Averages (EWMA)
eth_trend['EWMA12'] = eth_trend['price'].ewm(span=12, adjust=False).mean()
eth_trend['EWMA120'] = eth_trend['price'].ewm(span=120, adjust=False).mean()
eth_trend['EWMA240'] = eth_trend['price'].ewm(span=240, adjust=False).mean()
eth_trend['EWMA2400'] = eth_trend['price'].ewm(span=2400, adjust=False).mean()

# Plotting all trends
eth_trend.plot(x='TIME', y=[
    'price', '6-month-SMA', '12-month-SMA',
    '600-day-SMA', '1200-day-SMA',
    'EWMA12', 'EWMA120', 'EWMA240', 'EWMA2400'
], figsize=(14, 6))
plt.title('Ethereum Price Trend with SMA and EWMA Smoothing')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.show()
```

```python
eth_trend.rename(columns={'TIME': 'ds', 'price': 'y'}, inplace=True)
eth_trend

```

```python
from prophet import Prophet
```

```python
eth_trend['ds'] = eth_trend['ds'].dt.tz_localize(None)
```

```python
m=Prophet()
m.fit(eth_trend)
```

```python
future = m.make_future_dataframe(periods=200)  # Forecast 200 days ahead
future
```

```python
forecast = m.predict(future)
forecast
```

```python
forecast.tail()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(forecast['ds'], forecast['trend'], label='Trend', color='blue')
plt.title('Ethereum Price Trend Component')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
forecast_range = forecast[(forecast['ds'] >= '2025-08-09') & (forecast['ds'] <= '2025-09-01')]
print(forecast_range[['ds', 'yhat']])
```

```python
plt.figure(figsize=(12, 6))
plt.plot(forecast_range['ds'], forecast_range['yhat'], marker='o', linestyle='-', color='green')
plt.title('Predicted Ethereum Prices (Aug–Sep 2025)')
plt.xlabel('Date')
plt.ylabel('Predicted Price (USD)')
plt.grid(True)
plt.show()
```
